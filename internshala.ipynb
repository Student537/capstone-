```python
import pandas as pd

df = pd.read_csv('internshala.csv')

if 'stipend' in df.columns:
    df['stipend'] = df['stipend'].fillna(0)

```

```python
df.rename(columns={'profile': 'domain'}, inplace=True)
df.columns = df.columns.str.strip().str.lower()

plt.figure(figsize=(30,9))
sns.countplot(x='internship name', data=df)
plt.xticks(rotation=60)
plt.show()

```

```python
monthly_demand = df.groupby('stipend').size()

plt.figure(figsize=(25,4))
monthly_demand.plot(kind='line', marker='o')
plt.title('stipend Internship Demand Trend')
plt.xlabel('stipend')
plt.ylabel('Number of Internships')
plt.show()
```

```python
if 'company' in df.columns and 'internship name' in df.columns:
    plt.figure(figsize=(30,25))
    sns.barplot(x='internship name', y='company', data=df)
    plt.title('Applications per Internship')
    plt.xticks(rotation=60)
    plt.show()

```

```python
 if 'location' in df.columns:
    plt.figure(figsize=(8,4))
    df['location'].value_counts().head(10).plot(kind='bar')
    plt.title('Top 10 Internship Locations')
    plt.ylabel('Count')
    plt.show()
```
